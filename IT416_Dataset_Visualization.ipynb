```python
import os
import random
random.seed(42)
import cv2
import matplotlib.pyplot as plt
```

```python
print('Total distinct people:', len(os.listdir('processed_dataset')))
```

```python
print('5 random Face images\n')

people = os.listdir('processed_dataset')
random.shuffle(people)
people = people[: 5]

plt.rcParams['figure.figsize'] = (18, 5)
fig, ax = plt.subplots(1, 5)

for i, person in enumerate(people):
  folder = 'processed_dataset' + '/' + str(person) + '/' + 'face'
  files = os.listdir(folder)
  random.shuffle(files)
  file = files[0]

  path = folder + '/' + file
  image = cv2.imread(path)

  ax[i].imshow(image)

plt.show()
```

```python
print('Gait Energy Images of a person from 90: normal, holding a bag, wearing a coat\n')

people = os.listdir('processed_dataset')
random.shuffle(people)
person = people[0]

plt.rcParams['figure.figsize'] = (24, 8)
fig, ax = plt.subplots(3, 8)

styles = ['nm-01', 'bg-01', 'cl-01']

for i, style in enumerate(styles):
  folder = 'processed_dataset' + '/' + str(person) + '/' + 'gait' + '/' + style + '/' + '090'
  files = os.listdir(folder)
  random.shuffle(files)
  files = files[: 8]
  files.sort()

  for j, file in enumerate(files):
    path = folder + '/' + file
    image = cv2.imread(path)
    ax[i, j].imshow(image)
    ax[i, j].axis('off')

plt.show()
```
